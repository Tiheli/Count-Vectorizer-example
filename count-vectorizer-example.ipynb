```python
# Importing CountVectorizer to transform text data into numerical feature vectors for analysis
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Creating an empty DataFrame to store text data and associated labels
df = pd.DataFrame()
```

```python
# Populating the DataFrame with a sample list of phrases in the 'words' column for text analysis
df['words'] = ["ali look", "ali look the horse", "look ali look", "ali look the beauty horse"]
```

```python
# Initializing CountVectorizer to convert text data into a matrix of token counts
cv = CountVectorizer()
```

```python
# Fitting and transforming the 'words' column to produce a sparse matrix of token counts for each phrase
a = cv.fit_transform(df['words'])
```

```python
# Converting the sparse matrix to a dense array for easier viewing and manipulation
x = a.toarray()
```

```python
# Retrieving the feature names (unique words) extracted by CountVectorizer
cv.get_feature_names_out()
```
